```python


from google.colab import files
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
import zipfile
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier , AdaBoostClassifier
from sklearn.metrics import confusion_matrix , classification_report , accuracy_score , roc_auc_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
import shap
import eli5
from eli5.sklearn import PermutationImportance
from sklearn.neural_network import MLPClassifier
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from statsmodels.graphics.mosaicplot import mosaic


#Suppresses Warnings.

warnings.filterwarnings('ignore')

#Uploads the Zip File.
uploaded = files.upload()

#Extracts the Contents of the Zip File.
file_name = "Data.zip"
with zipfile.ZipFile(file_name, 'r') as zip_ref:
    zip_ref.extractall()
    print('done')

#Loads the CSV file.
heart = pd.read_csv('/content/Data/heart.csv')

#Converts Columns to Appropriate Data Types.
numeric_columns = ["age", "trestbps", "chol", "thalach", "oldpeak", "slope", "ca", "thal"]
for col in numeric_columns:
    heart[col] = pd.to_numeric(heart[col], errors='coerce')

#Handles Missing Values if they exist.
heart = heart.dropna()

#Renames our Columns/Variables
heart = heart.rename(columns={"cp": "chest_pain", "trestbps": "blood_pressure", "fbs": "blood_sugar", "ca": "vessels", "chol": "cholesterol"})
heart['health_status'] = heart['target']
heart['health_status'] = ["healthy" if x == 0 else "sick" for x in heart['health_status']]
heart['gender'] = heart['sex']
heart['gender'] = ['F' if x == 0 else 'M' for x in heart['gender']]

#Displays the First and Last few Rows.
print(heart.head())
print(heart.tail())

#Display Basic Information about the Dataset.
print(heart.shape)
print(heart.describe())
print(heart.dtypes)

#Checks for Duplicate Rows and Removes them.
duplicates = heart[heart.duplicated(keep=False)]
heart = heart.drop_duplicates(keep='first')
print(heart['health_status'].value_counts())


```

```python
'''
By Normalizing our Data, we can improve the Accuracy of our Models/Algorithms, prevent Bias from occuring and have Consistency across our Features.
'''
#Normalization of our Data.
#Standardizes All Numerical Features (Except the Target and One-Hot Encoded).
numerical_features = heart.select_dtypes(include=['number'])
numerical_features = numerical_features.drop(columns=['target'])

standardScaler = StandardScaler()
scaled_data = standardScaler.fit_transform(heart[numerical_features.columns])

#Creates a New DataFrame for our Standardized Numerical Features.
scaled_df = pd.DataFrame(scaled_data, columns=numerical_features.columns, index=heart.index)

#Joins the Scaled Numerical Features back to the Original DataFrame.
heart = heart.drop(numerical_features.columns, axis=1).join(scaled_df)
```

```python
'''
Using the 2 Links found on Lab 4 of our Module's Moodle Page, we explored the Topic of how we can split our Data into Training, Validation and
Testing Sets. In order for us to create our Model and put our Data through a Machine Learning Process, we need to know what Percentage
of our Data we need to Train and Test. For this, we need 2 Datasets, "X" which has Independent Features and "Y" which has Dependent Variables.
The Formula to split our Data is simply X Train - X Test / Y Train - Y Test. Once we create our Model, the Input/Output of our "X" Test,
should be equal to our "Y" Test. The closer to it, the better the Accuracy. To achieve this, we need to split our Data into 3 Sets, Training,
Validation and Testing. But we need to be wary of 2 Issues. Overfitting and Underfitting. Overfitting is when our Model has been "trained too well"
with our Data that it has now become too accustomed to them, making the Process of training New Untrained Data inaccurate. This can happen due to
our Model being too complex. However, Underfitting can occur when our Model is"too simple" and can't fit the Training Data. We need enough Independent
Variables but not too many, so we cna also avoid Overfitting. Finally, Validation is when we split our Data into "k" Subsets and train "k-1" of them.
We use the Last Subset for Testing. This can be achieved using Libraries such as Pandas, SKLearn or Matplotlib. Below this Comment, we have successfully
split our Dtata into Training and Testing Sets.
'''


#Encodes our Categorial Features.
heart = pd.get_dummies(heart, columns=['sex' , 'chest_pain' , 'blood_sugar' , 'restecg' , 'exang' , 'slope' , 'thal'], drop_first=True)

#Separates Features from Target Labels.
labels = heart['target']
features = heart.drop(['target', 'health_status', 'gender'], axis=1)

#Splits our Features and Target Labels into a Training Set and a Test Set.
features_train , features_temp, labels_train , labels_temp = train_test_split(features , labels , test_size=0.2 , random_state=42)
features_val, features_test, labels_val, labels_test = train_test_split(features_temp , labels_temp , test_size=0.5 , random_state=42)

#Displays the Shapes of the Resulting Datasets.
print("Training Set Shape: ", features_train.shape , labels_train.shape)
print("Testing Set Shape: ", features_test.shape , labels_test.shape)
print("Testing Set Shape: ", features_test.shape , labels_test.shape)
```

```python
'''
In order for Us to train our First Machine Learning Model, we need to use Models such as Logistic Regression, K-Nearest Neighbours, Random Forest
and Naive-Bayes. This is done after splitting our Data into Test and Train Sets, a Process we have already taken care of. We will first train our Model
using the Random Forest Model. It uses multiple Decision Trees(otherwise known as CARTs). The Individual Trees are built via Bagging and split using fewer Features.
The Result is a Diverse Forest of Uncorrelated Tree Exhibits, which reduce Variance and therefore making this Model more robust towards any Changes in Data, while
also carrying its Prediction Accuracy to any New Data. These are our Findings using this Model:
'''
#Trains and Evaluates our Random Forest Model with Different Number of Trees.
randomForest_scores = []
trees = [10 , 100 , 200 , 500 , 1000 , 1500 , 2000 , 5000]
for x in trees:
  randomForest = RandomForestClassifier(n_estimators=x ,random_state = 1 , max_depth = 1 )
  randomForest.fit(features_train , labels_train)
  randomForest_scores.append(randomForest.score(features_val , labels_val))

#Our Plot Accuracy VS Our Number Of Trees.
plt.figure(figsize = (10 , 6))
sns.barplot(x=trees , y=randomForest_scores , palette='Blues')
plt.xlabel('Number of Trees')
plt.ylabel('Accuracy Score')
plt.title('Accuracy VS Number of Trees')
plt.show()

#Trains and Evaluates our Random Forest Model with Different Max Depths.
randomForest_scores = []
depth = [1 , 5 , 10 , 15]
for x in depth:
  randomForest = RandomForestClassifier(n_estimators = 1000 , random_state = 42 , max_depth = x)
  randomForest.fit(features_train , labels_train)
  randomForest_scores.append(randomForest.score(features_val , labels_val))

#Our Plot Accuracy VS Max Depth.
plt.figure(figsize = (10 , 6))
sns.barplot(x=depth , y=randomForest_scores , palette='Blues')
plt.xlabel('Max Depth')
plt.ylabel('Accuracy Score')
plt.title('Accuracy VS Max Depth')
plt.show()

#Calculates our Accuracy.
score = round(randomForest.score(features_test , labels_test) , 3) * 100
print(f"Random Forest Accuracy: {score}%")

#Features Importance using Permutation Importance.
perm = PermutationImportance(randomForest , random_state = 42).fit(features_test , labels_test)
eli5.show_weights(perm , feature_names = features_test.columns.tolist())

#SHAP Values for Feature Importance.
explainer = shap.TreeExplainer(randomForest)
shap_values = explainer.shap_values(features_test)

#Summary Plot of our SHAP Values.
shap.summary_plot(shap_values[:, :, 1], features_test, plot_type="bar")

#Our Confusion Matrix.
labels_predicted = randomForest.predict(features_test)
plt.figure(figsize = (10 , 5))
conf_mat = confusion_matrix(labels_test , labels_predicted)
sns.heatmap(conf_mat , annot = True , fmt = 'd' , cmap = 'Blues' , cbar = False , xticklabels = ['healthy' , 'sick'] , yticklabels = ['healthy' , 'sick'])
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Random Forest: Confusion Matrix')
plt.show()
```

```python
'''
Now we will implement the K-Nearest Neighbor Model. It predicts Results based on the specified number "k" of the
Nearest Neighboring Data Points. The Pre-Processing of the Data is Significant, as it can impact the Distance Measurements.
Moreover, the "k" Parameter needs to be chosen wisely, as a Value Lower than Optimal can lead to Bias, whereas a
Higher Value can impact Prediction Accuracy. It is Simple to Implement and preferable for Smaller Datasets.
'''
#Initializes our K-Nearest Neighbor Model with 11 Neighbors.
knn = KNeighborsClassifier(n_neighbors = 11)
knn.fit(features_train , labels_train)

#Makes Predictions with our K-Nearest Neighbor Model.
prediction = knn.predict(features_test)

#Calculates and Prints our Accuracy of our K-Nearest Neighbor Model.
score = round(knn.score(features_test , labels_test) , 3) * 100
print(f"K Nearest Neighbors Accuracy is {score}%")

#Stores our Accuracy Scores for Different Values of k.
accuracyScores = []

#Iterates over a Range of k Values from 1 to 30 and Stores our Accuracy Scores.
for x in range(1 , 30):
  knn2 = KNeighborsClassifier(n_neighbors = x)
  knn2.fit(features_train , labels_train)
  accuracyScores.append(knn2.score(features_val , labels_val))

#Plots our Accuracy Scores for Dieffrent k Values.
sns.lineplot(x=range(1 , 30) , y=accuracyScores)
plt.xticks(np.arange(1 , 30 , 1))
plt.xlabel("k Value")
plt.ylabel("Accuracy Score")

#Finds and Prints the Best k Value and its Corresponding Accuracy Score.
best_k = accuracyScores.index(max(accuracyScores)) + 1
max_score = round((max(accuracyScores)*100) , 2)
print(f"Max k Nearest Neighbors Accuracy is {max_score}%")
print(f"Best k is {best_k}")

#Initializes the K-Nearest Neighbor Model with the Best k Value found.
knn = KNeighborsClassifier(n_neighbors = best_k)
knn.fit(features_train , labels_train)

#Makes Predictions with our K-Nearest Neighbor Model.
prediction = knn.predict(features_test)

#Calculates and Prints the Accuracy of our K-Nearest Neighbor Model.
score = round(knn.score(features_test , labels_test) , 3) * 100
print(f"K Nearest Neighbors Accuracy is {score}%")

#Features the Importance using Permuation Importance.
perm = PermutationImportance(knn , random_state = 1).fit(features_test , labels_test)
eli5.show_weights(perm , feature_names = features_test.columns.tolist())

#Makes Predictions with our K-Nearest Neighbor Model.
labels_predicted = knn.predict(features_test)
plt.subplots(figsize = (10 , 5))

#Our Confusion Matrix for our K-Nearest Neighbor Model.
conf_mat = confusion_matrix(labels_test , labels_predicted)
sns.heatmap(conf_mat , annot = True , fmt = 'd' , cmap = 'Blues' , cbar = False , xticklabels = ['healthy' , 'sick'] , yticklabels = ['healthy' , 'sick'])
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('K Nearest Neighbors: Confusion Matrix')
plt.show()

#Prints the Classification Report for our K-Nearest Neighbor Model.
print(classification_report(labels_test , labels_predicted , target_names = ['healthy' , 'sick']))
```

```python
'''
Next, we will implement the Naive-Bayes Model. It applies what is known as a Posterior Probability.
It does this by using the Bayes Theorem to do the Categorization on the Unstructured Data. It then
makes a Naive Assumption that the Predictors can be Independent, which may not be True. It
is also preferably used for Smaller Datasets. Here is hwo we implemented it:
'''
#Initializes and Trains our Naive-Bayes Model.
nb = GaussianNB()
nb.fit(features_train , labels_train)

#Calculates and Prints our Accuracy of our Naive-Bayes Model.
score = round(nb.score(features_test , labels_test) , 3) * 100
print(f"Niave Bayes Accuracy is {score}%")

#Makes Predictions with our Naive-Bayes Model.
labels_predicted - nb.predict(features_test)
plt.subplots(figsize = (10 , 5))

#Our Confusion Matrix for Naive Bayes Model.
conf_mat = confusion_matrix(labels_test , labels_predicted)
sns.heatmap(conf_mat , annot = True , fmt = 'd' , cmap = 'Blues' , cbar = False , xticklabels = ['healthy' , 'sick'] , yticklabels = ['healthy' , 'sick'])
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Naive Bayes: Confusion Matrix')
plt.show()
```

```python
'''
Finally, we have the Logistic Regression Model, which uses Regression to call-out the Contribution
of our Individual Predictors. It also uses Maximum Likelihood to fit a Sigmoid-Curve on the Target
Variable Distribution. Here is how we did it:
'''
#Initializes and Trains the Logistic Regression Model.
logisticRegression = LogisticRegression(solver='lbfgs')
logisticRegression.fit(features_train , labels_train)

#Calculates and Prints the Accuracy of the Logistic Regression Model.
score = round(logisticRegression.score(features_test , labels_test) , 3)
print(f"Logistic Regression Accuracy is {score}%")

#Features Importance using Permutation Importance.
perm = PermutationImportance(logisticRegression , random_state = 1).fit(features_test , labels_test)
eli5.show_weights(perm , feature_names = features_test.columns.tolist())

#Makes Predictions with the Logistic Regression Model.
labels_predicted = logisticRegression.predict(features_test)
plt.subplots(figsize = (10 , 5))

#Creates our Confusion Matrix for our Logistic Regression Model.
conf_mat = confusion_matrix(labels_test , labels_predicted)
sns.heatmap(conf_mat , annot = True , fmt = 'd' , cmap = 'Blues' , cbar = False , xticklabels = ['healthy' , 'sick'] , yticklabels = ['healthy' , 'sick'])
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Logistic Regression: Confusion Matrix')
plt.show()

#Prints the Classification report for the Logistic Regression Model.
print(classification_report(labels_test , labels_predicted , target_names = ['healthy' , 'sick']))
```

```python
'''
Here, we have decided to include another Model for our Training/Testing/Validation Sets. This Time we will be
using the Neural Networks Model. This Machine Learning Model, mimics the Complex Functions
of our Brains. It consists of Interconnected Nodes that process Data, learn Patterns and are mostly
used in Decision-Making. It works by exposingitself to a Simulated Dataset. It's Parameters are Updated
in response to New Data/Conditions. With each Update, the Network's Response evolves, allowing it to
adapt effectively. It also consists of 3 Layers, Input, Hidden and Output. The MLP Model of it, which will
be the one we will be using, uses Non-Linear Activation Functions to work. Overall, this Model is also frequently
used for the purpose of diagnosing Disease, which makes it incredibly relevant to our Project. This is how we
implemented it.
'''

#Defines and Initializes the Multi-Layer Perceptron (MLP) Classifier with specified Hidden Layer Size and Max Iterations.
mlp_model = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)

#Trains the MLP Model on the Training Dataset.
mlp_model.fit(features_train, labels_train)

#Calculates and Prints the Model's Accuracy on the Test Dataset.
score = round(mlp_model.score(features_test, labels_test), 3)
print(f"Neural Network (MLP) Accuracy: {score * 100}%")

#Features Importance using Permutation Importance
#Calculates the Importance of Features by measuring the Effect of Permuting each Feature on the Model's Accuracy.
perm = PermutationImportance(mlp_model, random_state=1).fit(features_test, labels_test)

#Displays the Feature Importance Scores.
eli5.show_weights(perm, feature_names=features_test.columns.tolist())

#Makes Predictions on the Test Dataset using the Trained MLP Model.
labels_predicted = mlp_model.predict(features_test)

#Creates a Plot with a specific Figure Size.
plt.subplots(figsize=(10, 5))

#Generates the Confusion Matrix to visualize the Performance of the Model in Classifying Healthy vs. Sick Patients.
conf_mat = confusion_matrix(labels_test, labels_predicted)

#Plots the Confusion Matrix using a Heatmap for better Visualization.
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['healthy', 'sick'], yticklabels=['healthy', 'sick'])
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Neural Network (MLP): Confusion Matrix')
plt.show()

#Prints the Classification Report showing Precision, Recall, and F1-Score for each Class.
print(classification_report(labels_test, labels_predicted, target_names=['healthy', 'sick']))
```

```python
'''
QDA is a Classification Algorithm that uses a Quadratic Decision Boundary to separate
Data Points into Different Classes. It estimates the Mean Marix for each Class of our
Training Data. It is efficient and flexible. It uses Gaussian Distributions and the
Parameters of a New Data Point to calculate its Probability, ultimately assigning it
to the Class with the Highest Probability. Finally, it gives us a more Complex Output,
thna other Models. This is how we implemented it:
'''

#Defines And Initializes The Quadratic Discriminant Analysis (QDA) Model.
qda_model = QuadraticDiscriminantAnalysis()

#Trains The QDA Model On The Training Dataset.
qda_model.fit(features_train, labels_train)

#Calculates And Print The Model's Accuracy On The Test Dataset.
score = round(qda_model.score(features_test, labels_test), 3)
print(f"Quadratic Discriminant Analysis (QDA) Accuracy: {score * 100}%")

#Features Importances Using Permutation Importance.
perm = PermutationImportance(qda_model, random_state=1).fit(features_test, labels_test)

#Displays The Feature Importance Scores.
eli5.show_weights(perm, feature_names=features_test.columns.tolist())

#Makes Predictions On The Test Dataset Using The Trained QDA Model.
labels_predicted = qda_model.predict(features_test)

#Creates A Plot With A Specific Figure Size.
plt.subplots(figsize=(10, 5))

#Confusion Matrix
conf_mat = confusion_matrix(labels_test, labels_predicted)

#Plots The Confusion Matrix Using A Heatmap For Better Visualization.
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['healthy', 'sick'], yticklabels=['healthy', 'sick'])
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Quadratic Discriminant Analysis (QDA): Confusion Matrix')
plt.show()

#Prints The Classification Report Showing Precision, Recall, And F1-Score For Each Class.
print(classification_report(labels_test, labels_predicted, target_names=['healthy', 'sick']))
```

```python
'''
AdaBoost is a Statistical Classification Meta-Algorithm. It can be used with
many other Models to improve Performance. the Output of Multiple Weak Learners
is combined into a Weighted Sum that represents the Final Output of the Boosted
Classifier. Finally, it can also be less prone to Overfitting. This is how we
implemented it:
'''

#Defines And Initializes The AdaBoost Classifier With Specified Parameters.
adaboost_model = AdaBoostClassifier(n_estimators=50, random_state=42)

#Train The AdaBoost Model On The Training Dataset.
adaboost_model.fit(features_train, labels_train)

#Calculates And Prints The Model's Accuracy On The Test Dataset.
score = round(adaboost_model.score(features_test, labels_test), 3)
print(f"AdaBoost Accuracy: {score * 100}%")

#Features Importance Using Permutation Importance.
perm = PermutationImportance(adaboost_model, random_state=1).fit(features_test, labels_test)

#Displays The Feature Importance Scores.
eli5.show_weights(perm, feature_names=features_test.columns.tolist())

#Makes Predictions On The Test Dataset Using The Trained AdaBoost Model.
labels_predicted = adaboost_model.predict(features_test)

#Creates A Plot With A Specific Figure Size.
plt.subplots(figsize=(10, 5))

#Creates our Confusion Matrix.
conf_mat = confusion_matrix(labels_test, labels_predicted)

#Plots The Confusion Matrix Using A Heatmap For Better Visualization.
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['healthy', 'sick'], yticklabels=['healthy', 'sick'])
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('AdaBoost: Confusion Matrix')
plt.show()

#Prints The Classification Report Showing Precision, Recall, And F1-Score For Each Class.
print(classification_report(labels_test, labels_predicted, target_names=['healthy', 'sick']))
```
